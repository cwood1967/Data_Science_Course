### Matplotlib
- simple plots with lists
- plot with markers

```python
### Matplotlib

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

### Just a simple plot
You can plot a simple list of numbers. When plotting a list like this, matplotlib assumes the list is the data for the y-axis.

```python
y = [2,4,1,6,7,3]
plt.plot(y)
```

### Make an y vs. x plot with two lists

```python
x = [1,2,5,7,9,10]
y = [2,4,1,6,7,3]
plt.plot(x, y)
```

## Line color and width
- color: color='blue' or c='blue'
- line: linewidth=X or lw=X  where x is a floating point number
- linestyle : linestyle=[ '-' | '--' | '-.' | ':' | 'steps' | ...]

```python
### show the data points and color red with 'ro-'
plt.plot(x, y, c='red', lw=3.4, linestyle='--')
```

## Plot markers

```python
s = plt.plot(x, y, c='orange', marker='o', markersize='12', linestyle='')
```

## Size of the plot

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, c='orange', marker='o', markersize='12', linestyle='')
```

## Size, Titles, and labels, figure size

```python
plt.plot(x, y, marker='o')
plt.title("Example plot from lists")
plt.xlabel("x values")
plt.ylabel("y values")
```
