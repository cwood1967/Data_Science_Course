```python
### Matplotlib

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

### Matplotlib

Matplotlib is the general python plotting library that other plotting libraries are based on (Seaborn, pandas plotting functions). It is good idea to use something like Seaborn to easily make great-looking plots with error bars and confidence intervals. Learning some matplotlib basics will let you quickly look at data and also use more options in seaborn.

Matplotlib can a few different sources of data: python lists, numpy arrays, and pandas Series, and dictionaries.

### Just a simple plot
You can plot a simple list of numbers. When plotting a list like this, matplotlib assumes the list is the data for the y-axis.

```python
y = [2,4,1,6,7,3]
plt.plot(y)
```

### Make an y vs. x plot with two lists

Two lists (of the same length) can be used with the plot command. The first is x, the second y. The lists don't need to be named x and y:

```python
time = [1,2,5,7,9,10]
distance = [2,2.2,8,9,12,13]
plt.plot(time, distance)
```

## Line color and width
- color: color='blue' or c='blue'
- line: linewidth=X or lw=X  where x is a floating point number
- linestyle : linestyle=[ '-' | '--' | '-.' | ':' | 'steps' | ...]

```python
### show the data points and color red with 'ro-'
plt.plot(time, distance, c='red', lw=3.4, linestyle='--')
```

## Plot markers
In the case of our x variable, time, the data points are not evenly distributed. Using markers will help visualize the time points.

```python
s = plt.plot(time, distance, c='orange', marker='o', markersize='12', linestyle='-')
```

## Size of the plot

The size of the plot can be changed using plt.figure:

```python
plt.figure(figsize=(10,5))
plt.plot(time, distance, c='orange', marker='o', markersize='12', linestyle='-')
```

## Size, Titles, and labels, figure size
The names of the axes can be set with `xlabel` and `ylabel`. Fonts can be changed with `fontsize` and `fontfamily`.

```python
plt.plot(time, distance, marker='o')
plt.title("Example plot from lists", fontsize=16, fontfamily='serif')
plt.xlabel("Time", fontsize=12)
plt.ylabel("Distance", fontsize=12)
```
