```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/richard-alexander/Data_Science_Course/master/Seaborn/SubmittedData/MCK_Data.csv')
```

```python
df['Filename'].unique()
```

```python
df['Condition'] = df['Filename'].str.split('Tracks').str[0]
```

```python
plt.figure(figsize=[15,8])
sns.barplot(data=df, x='Condition', y='Track Speed Max')
```

```python
potentially_interesting = ['Track Displacement Y', 'Track Displacement X',
       'Track Displacement Length', 'Track Duration', 'Track Length',
       'Track Position X Start', 'Track Speed Max', 'Track Speed Mean',
       'Track Speed Variation', 'Track Straightness', 'Alpha',
       'Y Start Center', 'Gamma', 'Intensity', 'Mean InstaSpeed',
       'Max Acceleration', 'Velocity X', 'Velocity Y', 'Time Interval',
       'X Start Center', ]
```

```python
for feature in potentially_interesting:
    plt.figure(figsize=[15,8])
    sns.violinplot(data=df, x='Condition', y=feature, order=['InVitro', 'PodCntl', 'PodXL', 'Ang2'])
    sns.swarmplot(data=df, x='Condition', y=feature, order=['InVitro', 'PodCntl', 'PodXL', 'Ang2'], dodge=True)
```
